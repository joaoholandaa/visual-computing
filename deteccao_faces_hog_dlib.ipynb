# Detecção HOG/DLIB

```python
import dlib
import cv2 
```

```python
imagem = cv2.imread('/content/people2.jpg')
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(imagem)
```

```python
detector_face_hog = dlib.get_frontal_face_detector()
```

```python
deteccoes = detector_face_hog(imagem, 4) #escala
```

```python
deteccoes, len(deteccoes)
```

```python
for face in deteccoes:
	l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
	cv2.rectangle(imagem, (l, t), (r, b), (0, 255, 255), 2)
cv2_imshow(imagem)
```

# Detecção com CNN's



```python
imagem2 = cv2.imread('/content/people2.jpg')
detector_face_cnn = dlib.cnn_face_detection_model_v1('/content/mmod_human_face_detector.dat')
```

```python
deteccoes = detector_face_cnn(imagem2, 1)
for face in deteccoes:
	l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
	cv2.rectangle(imagem2, (l,t), (r,b), (255, 255, 0), 2)
cv2_imshow(imagem2)
```
